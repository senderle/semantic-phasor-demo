```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, TapTool, PointDrawTool, OpenURL, ColumnDataSource
from bokeh.palettes import magma

output_notebook()
```

```python

source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5],
    y=[2, 5, 8, 2, 7],
    desc=['A', 'b', 'C', 'd', 'E'],
))

TOOLTIPS = [
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("desc", "@desc"),
]

p = figure(plot_width=400, plot_height=400,
           title="Mouse over the dots",
           # tooltips=TOOLTIPS,
           tools='box_zoom, reset', 
           )


c1 = p.circle('x', 'y', size=7, source=source)
point_draw = PointDrawTool(renderers=[c1])
p.add_tools(point_draw)
p.toolbar.active_drag = point_draw

show(p)
```
